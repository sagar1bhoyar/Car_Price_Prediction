```python
price['CarName'].astype('category').value_counts()
```

```python
price['CarName'] = price['CarName'].apply(lambda x : x.split(" ")[0])
price['CarName'].value_counts()
```

```python
price.loc[(price['CarName'] == 'vokswagen') | 
                             (price['CarName'] == 'vw') 
                              , 'CarName'] = 'volkswagen'

price.loc[price['CarName'] == 'porcshce' , 'CarName'] = 'porsche'

price.loc[price['CarName'] == 'toyouta' , 'CarName'] = 'toyota'

price.loc[price['CarName'] == 'Nissan', 'CarName'] = 'nissan'

price.loc[price['CarName'] == 'maxda' , 'CarName'] = 'mazda'
```

```python
price['CarName'].value_counts()
```

```python
cars_categorical = price.select_dtypes('object')
price = price.drop(cars_categorical , axis=1)
price.head()
```

```python
Y = price['price']
X = price.drop(['price'] , axis=1)
```

```python
from sklearn.preprocessing import scale
cols = X.columns
X = pd.DataFrame(scale(X))
X.columns = cols
X.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test, Y_train, Y_test = train_test_split(X,Y, train_size = 0.7, test_size = 0.3)                                                     
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
Y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(r2_score(Y_pred, Y_test))
```
